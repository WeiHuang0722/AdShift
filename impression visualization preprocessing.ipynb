```python
demo = pd.read_csv('demo.csv')
demo
```

```python
demo_pop = demo['count'].sum()
demo_pop_by_zip = demo.groupby(['zipcode']).agg({'count': 'sum'})
demo_pop_by_zip.rename(columns={"count": "household_count"}, inplace=True)
demo_pop_by_zip['household_count'] = demo_pop_by_zip['household_count'].astype(float)
demo_pop_by_zip['expected_zip_freq'] = demo_pop_by_zip['household_count'] / demo_pop
demo_pop_by_zip.head()
```

```python
imp_by_zip = pd.DataFrame(data_clean['zip_code'].value_counts()).reset_index().rename(columns={"zip_code": "impression_count", "index": "zip_code"})
imp_by_zip['zip_code'] = imp_by_zip['zip_code'].astype(int)
imp_by_zip.set_index('zip_code', inplace=True)
imp_by_zip['observed_zip_freq'] = data_clean['zip_code'].value_counts(normalize=True)
imp_by_zip.head()
```

```python
pd.concat([imp_by_zip, demo_pop_by_zip], axis=1, join='outer', sort=False)
```

```python
joined_for_measurement = imp_by_zip.join(demo_pop_by_zip).dropna()
joined_for_measurement
```

```python
total_no = len(data_clean)

joined_for_measurement['expected count'] = joined_for_measurement['expected_zip_freq']*total_no
joined_for_measurement['observed count'] = joined_for_measurement['observed_zip_freq']*total_no

E_sub_O_sq = (joined_for_measurement['observed count'].dropna() - joined_for_measurement['expected count'].dropna())**2
chisq_rs = sum(E_sub_O_sq/joined_for_measurement['expected count'])
chisq_rs
```

```python

region = data_clean[['region', 'zip_code']].drop_duplicates()
map_zip_to_region = region.set_index('zip_code').to_dict().get('region')
```

```python
demo['region'] = demo['zipcode'].map(map_zip_to_region)
demo.dropna()
```

```python
imp_region = pd.DataFrame(data_clean['region'].value_counts(normalize=True)).reset_index()
imp_region.rename(columns={"index": "region", "region": "imp_region_frequency"}, inplace=True)
imp_region

##question here: the joined table matches where the zip codes are the same... should the proportion in the demographic data and imp zip code freuqncy be adjusted so that it is the proportion just including the zip codes where they match?
demo_region = pd.DataFrame(demo['region'].value_counts(normalize=True)).reset_index()
demo_region.rename(columns={"index": "region", "region": "demo_region_frequency"}, inplace=True)

imp_region.set_index('region', inplace=True)
demo_region.set_index('region', inplace=True)

joined = imp_region.join(demo_region).dropna()
joined
```

```python
total_no = len(data_clean)

joined['expected freq'] = joined['demo_region_frequency']*total_no
joined['observed freq'] = data_clean['region'].value_counts()

E_sub_O_sq = (joined['observed freq'].dropna() - joined['expected freq'].dropna())**2
chisq_rs = sum(E_sub_O_sq/joined['expected freq'])
chisq_rs
```
